```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/TheTitanic.csv')
```

```python
df.to_csv('TheTitanic.csv')
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
df.replace("?", np.nan,inplace=True)
df.replace("**", np.nan, inplace=True)
df.head(25)
```

```python
for column in df:
  print(column)
  print(missing_data[column].value_counts())
  print("")
```

```python
hello = df.dropna()
hello.head(15)
```
